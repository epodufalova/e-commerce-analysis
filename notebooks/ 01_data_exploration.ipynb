```python
orders = pd.read_sql("SELECT * FROM orders;", conn)
customers = pd.read_sql("SELECT * FROM customers;", conn)
products = pd.read_sql("SELECT * FROM products;", conn)
order_items = pd.read_sql("SELECT * FROM order_items;", conn)
```

```python
for name, df in {"orders": orders, 
                 "customers": customers, 
                 "products": products, 
                 "order_items": order_items}.items():
    print(f"\n{name.upper()}")
    print("Shape:", df.shape)
    print(df.info())
```

```python
print("\nOrders sample:\n", orders.head(10))
print("\nCustomers sample:\n", customers.head(10))
print("\nProducts sample:\n", products.head(10))
print("\nOrder Items sample:\n", order_items.head(10))
```

```python
def missing_report(df):
    report = df.isnull().sum().reset_index()
    report.columns = ["column", "missing_count"]
    report["missing_pct"] = 100 * report["missing_count"] / len(df)
    return report


print("\nMissing values in ORDERS:\n", missing_report(orders))
print("\nMissing values in CUSTOMERS:\n", missing_report(customers))
print("\nMissing values in PRODUCTS:\n", missing_report(products))
print("\nMissing values in ORDER_ITEMS:\n", missing_report(order_items))
```

```python
total_orders = orders.shape[0]
total_customers = customers["customer_id"].nunique()
date_range = (orders["order_purchase_timestamp"].min(), 
              orders["order_purchase_timestamp"].max())
total_revenue = order_items["price"].sum()

print("\nBASIC STATISTICS")
print("Total orders:", total_orders)
print("Total customers:", total_customers)
print("Date range:", date_range)
print("Total revenue: $", total_revenue)
```
